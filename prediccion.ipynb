```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error, r2_score


df = pd.read_csv('housing.csv')
print(df)
print(df.info())

print('primeras filas del dataset')
print(df.head())

print('cantidad e filas con valores nulos')
print(df.isnull().sum())

print('estadisticas descriptivas')
print(df.describe())


print('correlacion matriz')

#seleccionar solo las columnas numericas
data_numerico = df.select_dtypes(include=['number'])

#calcular las matriz correlacion para las columnas numericas
correlacion_matrix = data_numerico.corr()
print(correlacion_matrix) 

x = df[["median_income","housing_median_age","total_rooms"]]
y = df[["median_house_value"]] 

x_train, x_test, y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=42) 

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
r2 = r2_score(y_test,y_pred)

print("R-Squared:", r2)
```
